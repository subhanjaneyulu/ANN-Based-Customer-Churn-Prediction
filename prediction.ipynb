```python
import tensorflow as tf 

import pickle
import pandas as pd 
import numpy as np 
```

```python
from tensorflow.keras.models import load_model
##load the trained models,scaler pickle,onehotencoder 
model=load_model('model.h5')
```

```python
import pickle
#load the encoder and scaler 
with open('onehot_encoder_geo.pkl','rb') as file:
    onehot_encoder_geo = pickle.load(file)
with open('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender = pickle.load(file)
with open("scaler",'rb') as file:
    scaler=pickle.load(file)
```

```python
import pandas as pd
##example input_data 
input_data = {
    "CreditScore":600,
    "Geography" :'France',
    "Gender" : 'Male',
    'Age':40,
    "Tenure":3,
    'Balance' : 60000,
    "NumOfProducts" : 2,
    'HasCrCard':1,
    "IsActiveMember":1,
    'EstimatedSalary':50000
}

#onehot encoding 
geo_encoder = onehot_encoder_geo.transform([[input_data['Geography']]])
geo_encoder_df = pd.DataFrame(geo_encoder.toarray(),columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
geo_encoder_df
```

```python
input_df = pd.DataFrame([input_data])
input_df
```

```python
#encode categorical gender 
input_df['Gender'] = label_encoder_gender.transform(input_df['Gender'])
input_df
```

```python
#concatination one hot encoder
input_df = pd.concat([input_df.drop('Geography', axis=1),geo_encoder_df],axis=1)
input_df
```

```python
###scaling the input data 
input_scaled = scaler.transform(input_df)
input_scaled
```

```python
#Prediction churn 
prediction=model.predict(input_scaled)
prediction
```
